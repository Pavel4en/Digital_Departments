# Линейная регрессия с z-стандартизацией

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm

# Загрузка данных из файла
data = pd.read_excel("merge_data.xlsx")

# Определение зависимой переменной и независимых переменных
dependent_variable = 'Итог_ВРП'
independent_variables = data.columns.drop(['Итог_ВРП', 'Год', 'Субъект', 'Итог_Врп_на_душу', 'Округ'])

# Z-стандартизация независимых переменных
X = data[independent_variables]
X = (X - X.mean()) / X.std()

# Выполнение степ-регрессии
y = data[dependent_variable]
X = sm.add_constant(X)  # Добавление константы (свободного члена)
model = sm.OLS(y, X).fit()

# Вывод результатов регрессии
print(model.summary())
```

# Z-стандартизация Степ-регрессия без округов

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression

# Загрузка данных
data = pd.read_excel("merge_data.xlsx")

# Определение зависимой переменной и независимых переменных
dependent_variable = 'Итог_ВРП'
independent_variables = data.columns.drop(['Итог_ВРП','Год','Субъект','Итог_Врп_на_душу','Округ'])

# Выделение независимых и зависимых переменных
X = data[independent_variables]
y = data[dependent_variable]

# Z-стандартизация независимых переменных
X = (X - X.mean()) / X.std()

# Создание объекта модели линейной регрессии
model = LinearRegression()

# Создание объекта для шаговой регрессии
rfe = RFE(estimator=model)

# Выполнение шаговой регрессии
rfe.fit(X, y)
selected_variables = np.array(independent_variables)[rfe.support_]

X_selected = sm.add_constant(data[selected_variables])
model_selected = sm.OLS(y, X_selected).fit()

print(model_selected.summary())
```
